```python
import pandas as pd

def remove_second_column(file_path, output_path):
    # Load the CSV file
    df = pd.read_csv(file_path)

    # Check if the DataFrame has at least two columns
    if df.shape[1] > 1:
        # Drop the second column (index 1)
        df_cleaned = df.drop(df.columns[1], axis=1)

        # Save the updated DataFrame to a new CSV file
        df_cleaned.to_csv(output_path, index=False)

        print(f"Second column removed. Updated file saved as '{output_path}'.")
    else:
        print("The file does not have a second column.")

# Replace '/content/merged_file_with_numbers.csv' with the path to your CSV file
# Replace '/content/cleaned_file.csv' with the desired path for the output file
remove_second_column('/content/merged_file_with_numbers.csv', '/content/cleaned_file.csv')
```
