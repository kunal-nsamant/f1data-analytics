```python
from bs4 import BeautifulSoup
import requests
```

```python
url = 'https://www.formula1.com/en/results.html/2023/races.html'

page = requests.get(url)

soup = BeautifulSoup(page.text, 'html')
```

```python
print(soup)
```

```python
table = soup.find_all('table')
```

```python
print(table)
```

```python
world_titles = soup.find_all('th')
```

```python
world_titles
```

```python
world_table_titles = [title.text for title in world_titles]
del world_table_titles[0]
del world_table_titles[6]
print(world_table_titles)
```

```python
import pandas as pd
df = pd.DataFrame(columns = world_table_titles)
df
```

```python
soup.find_all('tr')
```

```python
column_data = soup.find_all('tr')
```

```python
for row in column_data:
  row_data = row.find_all('td')
  individual_row_data = [data.text.strip() for data in row_data]
  print(individual_row_data)
```

```python
for row in column_data[1:]:
    row_data = row.find_all('td')
    individual_row_data = [data.text.strip() for data in row_data]

    if len(individual_row_data) > 1:  # Ensure there are enough elements
        individual_row_data.pop(-1)  # Remove last element (7th index)
        individual_row_data.pop(0)   # Remove first element (0th index)
        individual_row_data[2] = individual_row_data[2].replace("\n", " ")

    print(individual_row_data)
```
